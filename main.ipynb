```python
# importar as libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cria um df para trabalhar com os dados, usando apenas as colunas que interessam
df = df_original[[
    'NO_REGIAO',
    'SG_UF',
    'NO_MUNICIPIO',
    'NO_CATEGORIA',
    'NO_DEPENDENCIA',
    'CRE_CAT_0',
    'PRE_CAT_0',
    'FUN_01_CAT_0',
    'FUN_02_CAT_0',
    'FUN_03_CAT_0',
    'FUN_04_CAT_0',
    'FUN_05_CAT_0',
    'FUN_06_CAT_0',
    'FUN_07_CAT_0',
    'FUN_08_CAT_0',
    'FUN_09_CAT_0',
    'MED_01_CAT_0',
    'MED_02_CAT_0',
    'MED_03_CAT_0'
  ]]
```

### tratamento de valores nulos

```python
# verifica se existe algum valor nulo no df antes do tratamento
df.isnull().sum()
```

```python
# remove as colunas on NO_REGIAO Ã© nulo

# isso aconteceu pois ao final do arquivo existem linhas com comentÃ¡rios, porem ao remover as colunas desnecessÃ¡rias essas linhas ficaram com valores nulos

df = df[df['NO_REGIAO'].notnull()]
```

```python
# verifica novamente os valores nulos, afim de confirmar que o tratamento foi feito corretamente
df.isnull().sum()
```

```python
# corrige o restante das colunas que estÃ£o com valores nulos, substituindo por 0

# ao analisar a base antes, percebi que essas colunas nulas podem sim ter valor 0, pois sÃ£o valores de quantidade de horas, e como algumas escolas nÃ£o possuem determinadas series escolares, nÃ£o foi passado nenhum valor, o que significa que o valor Ã© 0

colunas_para_preencher_com_zero = [
    'CRE_CAT_0',
    'PRE_CAT_0',
    'FUN_01_CAT_0',
    'FUN_02_CAT_0',
    'FUN_03_CAT_0',
    'FUN_04_CAT_0',
    'FUN_05_CAT_0',
    'FUN_06_CAT_0',
    'FUN_07_CAT_0',
    'FUN_08_CAT_0',
    'FUN_09_CAT_0',
    'MED_01_CAT_0',
    'MED_02_CAT_0',
    'MED_03_CAT_0'
  ]

df[colunas_para_preencher_com_zero] = df[colunas_para_preencher_com_zero].fillna(0)
```

```python
# verifica novamente os valores nulos, afim de confirmar se a substituiÃ§Ã£o dos valores nulos por 0 foi feito corretamente
df.isnull().sum()
```

### remover os registros de 'Total'

```python
# removendo 'Total' da coluna 'NO_DEPENDENCIA'

df = df[df['NO_DEPENDENCIA'] != 'Total']
```

### grupos

```python
# contagem municipios por uf

contagem_municipios_por_uf = df.groupby('SG_UF')['NO_MUNICIPIO'].nunique()

contagem_municipios_por_uf
```

```python
def calcular_media_ignorando_zero(series):
    """Calcula a mÃ©dia de uma Series, ignorando os valores iguais a zero."""
    valores_validos = series[series != 0]
    if not valores_validos.empty:
        return valores_validos.mean()
    else:
        return 0
```

```python
# media de horas por dependÃªncia

media_horas_por_dependencia = df.groupby('NO_DEPENDENCIA')[[
    'CRE_CAT_0',
    'PRE_CAT_0',
    'FUN_01_CAT_0',
    'FUN_02_CAT_0',
    'FUN_03_CAT_0',
    'FUN_04_CAT_0',
    'FUN_05_CAT_0',
    'FUN_06_CAT_0',
    'FUN_07_CAT_0',
    'FUN_08_CAT_0',
    'FUN_09_CAT_0',
    'MED_01_CAT_0',
    'MED_02_CAT_0',
    'MED_03_CAT_0'
  ]].apply(calcular_media_ignorando_zero)

media_horas_por_dependencia
```

```python
# media de horas por UF

media_horas_por_uf = df.groupby('SG_UF')[[
    'CRE_CAT_0',
    'PRE_CAT_0',
    'FUN_01_CAT_0',
    'FUN_02_CAT_0',
    'FUN_03_CAT_0',
    'FUN_04_CAT_0',
    'FUN_05_CAT_0',
    'FUN_06_CAT_0',
    'FUN_07_CAT_0',
    'FUN_08_CAT_0',
    'FUN_09_CAT_0',
    'MED_01_CAT_0',
    'MED_02_CAT_0',
    'MED_03_CAT_0'
  ]].apply(calcular_media_ignorando_zero)

media_horas_por_uf
```

```python
# media de horas por regiÃ£o

media_horas_por_regiao = df.groupby('NO_REGIAO')[[
    'CRE_CAT_0',
    'PRE_CAT_0',
    'FUN_01_CAT_0',
    'FUN_02_CAT_0',
    'FUN_03_CAT_0',
    'FUN_04_CAT_0',
    'FUN_05_CAT_0',
    'FUN_06_CAT_0',
    'FUN_07_CAT_0',
    'FUN_08_CAT_0',
    'FUN_09_CAT_0',
    'MED_01_CAT_0',
    'MED_02_CAT_0',
    'MED_03_CAT_0'
  ]].apply(calcular_media_ignorando_zero)

media_horas_por_regiao
```

### graficos

```python
# Resetando o Ã­ndice para usar 'NO_REGIAO' como coluna
media_horas_por_regiao = media_horas_por_regiao.reset_index()

# Transformando o DataFrame para o formato longo
media_horas_por_regiao_melted = media_horas_por_regiao.melt(
    id_vars='NO_REGIAO',
    value_vars=[
        'CRE_CAT_0', 'PRE_CAT_0', 'FUN_01_CAT_0', 'FUN_02_CAT_0',
        'FUN_03_CAT_0', 'FUN_04_CAT_0', 'FUN_05_CAT_0', 'FUN_06_CAT_0',
        'FUN_07_CAT_0', 'FUN_08_CAT_0', 'FUN_09_CAT_0', 'MED_01_CAT_0',
        'MED_02_CAT_0', 'MED_03_CAT_0'
    ],
    var_name='Nivel_Ensino',
    value_name='Media_Horas'
)

# Criando o grÃ¡fico de barras agrupado
plt.figure(figsize=(16, 8))
sns.barplot(
    x='NO_REGIAO',
    y='Media_Horas',
    hue='Nivel_Ensino',
    data=media_horas_por_regiao_melted
)
plt.title('MÃ©dia de Horas-Aula por RegiÃ£o e NÃ­vel de Ensino (Ignorando Zeros)')
plt.xlabel('RegiÃ£o')
plt.ylabel('MÃ©dia de Horas-Aula DiÃ¡ria')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
